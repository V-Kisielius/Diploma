```python
# take all *.abz files from data folder and create a list of SynopticMap.filaments.shape for each of them
# then create a list of all unique shapes

import os
import numpy as np

path_to_data = './Kislovodsk/'
abz_files = [f for f in os.listdir(path_to_data) if f.endswith('.abz')]
abz_files.sort()

shapes = []
for f in abz_files:
    mymap = SynopticMap(path_to_data + f)
    shapes.append(mymap.filaments.shape)

unique_shapes = np.unique(shapes, axis=0)


print(f'x from {min(unique_shapes[:, 0])} to {max(unique_shapes[:, 0])}')
print(f'y from {min(unique_shapes[:, 1])} to {max(unique_shapes[:, 1])}')
```

```python
import os
from synmap import SynopticMap
import cv2
import numpy as np
from matplotlib import pyplot as plt
path_to_data = './data/'
```

```python
synmaps = []
for f in os.listdir(path_to_data):
    if f.endswith('.abz'):
        mymap = SynopticMap(path_to_data + f)
        synmaps.append(cv2.resize(mymap.filaments, (200, 100)))
synmaps = np.array(synmaps)
```

```python
for i, synmap in enumerate(synmaps):
    plt.imshow(synmap, cmap='gray')
    plt.title(1904+i)
    plt.show()
```

```python
import torch

synmaps = torch.tensor(synmaps)
torch.save(synmaps, 'Kislovodsk.pt')
```
