# Kislovodsk

```python
import os
from scripts.kislovodsk import KislovodskMap
import cv2
import numpy as np
from matplotlib import pyplot as plt
path_to_data = './Kislovodsk/'
```

```python
synmaps = []
for f in os.listdir(path_to_data):
    if f.endswith('.abz'):
        mymap = KislovodskMap(path_to_data + f)
        synmaps.append(cv2.resize(mymap.filaments, (200, 100)))
synmaps = np.array(synmaps)
for i, synmap in enumerate(synmaps):
    plt.imshow(synmap, cmap='gray')
    plt.title(1904+i)
    plt.show()
```

```python
import torch

synmaps = torch.tensor(synmaps)
torch.save(synmaps, 'Kislovodsk.pt')
```
